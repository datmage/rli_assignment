## Proof of concept - Risk by Group
### QUESTION:
Which customers or policy groups may represent elevated financial risk? 
### INTERPRETATION:
What is a policy group? This is different than a policy_type?
The only other pieces of data I have are state and industry, so I will cross those data points together to make visuals for all combinations.


### By Customer

```
%%sql -r customer_risk
select
    top 10
    customer_id,
    customer_name,
    sum(payments_endorsements) as t_payments_endorsements,
    sum(claims) as t_claims,
    case when sum(payments_endorsements) <> 0 then sum(claims) / sum(payments_endorsements) end as loss_ratio
from rli.present.prs_customer_risk
group by
    customer_id,
    customer_name
having (t_claims > 0 or t_payments_endorsements > 0)
order by
    loss_ratio desc;
```

```
import matplotlib.pyplot as plt

df = customer_risk

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(df['CUSTOMER_NAME'], df['LOSS_RATIO'].astype(float), color='steelblue')
ax.set_xlabel('Loss Ratio')
ax.set_ylabel('Customer')
ax.set_title('Loss Ratio by Customer')
ax.axvline(x=1, color='red', linestyle='--', label='Break-even (1.0)')
ax.legend()
plt.tight_layout()
plt.show()
```
