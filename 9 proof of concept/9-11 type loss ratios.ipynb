## Proof of concept - Loss Ratios
### QUESTION:
Which policy types are generating the highest loss ratios?
### INTERPRETATION
Literal. View the overall loss ratios by policy type across the entire timeframe of the data set.



```
%%sql -r loss_ratio_data
select * from rli.present.prs_loss_ratio_by_policy_type;
```

```

import matplotlib.pyplot as plt

df = loss_ratio_data
print(df.to_string(index=False))

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(df['POLICY_TYPE'], df['LOSS_RATIO'])
ax.set_xlabel('Policy Type')
ax.set_ylabel('Loss Ratio')
ax.set_title('Loss Ratio by Policy Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
