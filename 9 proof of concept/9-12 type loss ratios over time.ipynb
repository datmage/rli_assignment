## Proof of concept - Loss Ratios (v2)
### QUESTION:
Which policy types are generating the highest loss ratios?
### INTERPRETATION
Time based. Overall loss ratios per policy type by month for 2025.

```
%%sql -r loss_ratio_data
select * from rli.present.prs_loss_ratio_by_policy_type_yearmonth where yearmonth like '2025__';
```

```
import matplotlib.pyplot as plt

df = loss_ratio_data
policy_types = sorted(df['POLICY_TYPE'].unique())
n = len(policy_types)

fig, axes = plt.subplots(n, 1, figsize=(12, 4 * n), sharex=True)
fig.suptitle('Loss Ratio by Policy Type Over Time 2025', fontsize=14, fontweight='bold')

for i, policy_type in enumerate(policy_types):
    ax = axes[i]
    group = df[df['POLICY_TYPE'] == policy_type].sort_values('YEARMONTH')
    ax.plot(group['YEARMONTH'].astype(str), group['LOSS_RATIO'], marker='o', color=f'C{i}')
    ax.set_ylabel('Loss Ratio')
    ax.set_title(policy_type)
    ax.axhline(y=1.0, color='black', linestyle='--', alpha=0.25)

axes[-1].set_xlabel('Year-Month')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### NOTES
Added a line at 1.0 to show where income loss happens within a month.
